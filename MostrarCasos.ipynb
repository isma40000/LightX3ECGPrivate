## Bibliotecas

```python
import numpy as np
import ecg_plot
import matplotlib.pyplot as plt
import pandas as pd
```

## Función para mostar un caso

args: 
- filename: nombre del fichero
- source: localización del fichero
- sample: sample de los datos
- factor: valor por el que se multiplican los datos

```python
def ecgPlot(filename,source,sample, factor = 1):
    data = np.load(source)
    data = data* factor
    #print(data)
    xml_leads = ['DI', 'DII', 'DIII', 'AVR', 'AVL', 'AVF', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6']
    ecg_plot.plot(data, sample_rate= sample,lead_index=xml_leads, title=f"{filename}")
```
